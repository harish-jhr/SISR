```python
import gdown
import zipfile
from pathlib import Path
import matplotlib.pyplot as plt
from PIL import Image
import random

# Drive
file_id = '112dNDsImy_gBgpGc91rIvvUm5RVOLp0C'
dest = 'dataset.zip'
gdown.download(f'https://drive.google.com/uc?id={file_id}', dest, quiet=False)

# Extraction
with zipfile.ZipFile(dest, 'r') as zip_ref:
    zip_ref.extractall()
```

```python
base_path = Path('dir/SISR/data/image_super_resolution_dataset/scaling_4x')
for split in ['train', 'test']:
    lr_folder = base_path / split / 'LR'
    hr_folder = base_path / split / 'HR'
    lr_images = list(lr_folder.glob('*.png'))
    hr_images = list(hr_folder.glob('*.png'))
    sample_indices = random.sample(range(len(lr_images)), 3)
    
    for idx in sample_indices:
        lr_img = Image.open(lr_images[idx])
        hr_img = Image.open(hr_images[idx])

        fig, axs = plt.subplots(1, 2, figsize=(8, 4))
        axs[0].imshow(lr_img)
        axs[0].set_title(f'{split.upper()} Low-Res')
        axs[0].axis('off')
        axs[1].imshow(hr_img)
        axs[1].set_title(f'{split.upper()} High-Res')
        axs[1].axis('off')
        plt.show()
```
