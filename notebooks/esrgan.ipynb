```python
#Imports
import os, cv2, random
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import models

from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim

# Paths
train_lr = '/ssd_scratch/harish.j/SISR/data/image_super_resolution_dataset/scaling_4x/train/LR'
train_hr = '/ssd_scratch/harish.j/SISR/data/image_super_resolution_dataset/scaling_4x/train/HR'
test_lr = '/ssd_scratch/harish.j/SISR/data/image_super_resolution_dataset/scaling_4x/test/LR'
test_hr = '/ssd_scratch/harish.j/SISR/data/image_super_resolution_dataset/scaling_4x/test/HR'
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Data
class SRDataset(Dataset):
    def __init__(self, lr_dir, hr_dir, patch_size=48, scale=4, patches_per_image=10):
        self.lr_paths = sorted([os.path.join(lr_dir, f) for f in os.listdir(lr_dir)])
        self.hr_paths = sorted([os.path.join(hr_dir, f) for f in os.listdir(hr_dir)])
        self.ps = patch_size
        self.scale = scale
        self.ppi = patches_per_image
        self.lr_patches, self.hr_patches = self.extract_patches()

    def extract_patches(self):
        lr_patches, hr_patches = [], []
        for lr_path, hr_path in zip(self.lr_paths, self.hr_paths):
            lr = cv2.imread(lr_path).astype(np.float32) / 255.
            hr = cv2.imread(hr_path).astype(np.float32) / 255.
            h, w = lr.shape[:2]
            for _ in range(self.ppi):
                x = random.randint(0, h - self.ps)
                y = random.randint(0, w - self.ps)
                lr_p = lr[x:x+self.ps, y:y+self.ps]
                hr_p = hr[x*self.scale:(x+self.ps)*self.scale, y*self.scale:(y+self.ps)*self.scale]
                if hr_p.shape[:2] != (self.ps*self.scale, self.ps*self.scale): continue
                lr_patches.append(np.transpose(lr_p, (2, 0, 1)))
                hr_patches.append(np.transpose(hr_p, (2, 0, 1)))
        return lr_patches, hr_patches

    def __len__(self): return len(self.lr_patches)

    def __getitem__(self, idx):
        return (torch.tensor(self.lr_patches[idx], dtype=torch.float32),
                torch.tensor(self.hr_patches[idx], dtype=torch.float32))

#Geberator
class ResidualDenseBlock(nn.Module):
    def __init__(self, channels):
        super().__init__()
        self.conv1 = nn.Conv2d(channels, channels, 3, padding=1)
        self.conv2 = nn.Conv2d(channels, channels, 3, padding=1)
        self.conv3 = nn.Conv2d(channels, channels, 3, padding=1)

    def forward(self, x):
        out = F.relu(self.conv1(x))
        out = F.relu(self.conv2(out))
        out = self.conv3(out)
        return x + out * 0.2

class Generator(nn.Module):
    def __init__(self, scale=4, num_blocks=5, nf=64):
        super().__init__()
        self.head = nn.Conv2d(3, nf, 3, padding=1)
        self.body = nn.Sequential(*[ResidualDenseBlock(nf) for _ in range(num_blocks)])
        self.upsample = nn.Sequential(
            nn.Conv2d(nf, nf * (scale ** 2), 3, padding=1),
            nn.PixelShuffle(scale),
            nn.Conv2d(nf, 3, 3, padding=1)
        )

    def forward(self, x):
        x = self.head(x)
        x = self.body(x)
        x = self.upsample(x)
        return torch.clamp(x, 0, 1)

#Discriminator
class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        def block(in_c, out_c, stride):
            return nn.Sequential(
                nn.Conv2d(in_c, out_c, 3, stride, 1),
                nn.BatchNorm2d(out_c),
                nn.LeakyReLU(0.2, inplace=True)
            )
        self.model = nn.Sequential(
            block(3, 64, 1),
            block(64, 64, 2),
            block(64, 128, 1),
            block(128, 128, 2),
            block(128, 256, 1),
            block(256, 256, 2),
            nn.AdaptiveAvgPool2d(1),
            nn.Flatten(),
            nn.Linear(256, 1)
        )

    def forward(self, x): return self.model(x)

#Perp-Loss
class VGGPerceptualLoss(nn.Module):
    def __init__(self):
        super().__init__()
        vgg = models.vgg19(pretrained=True).features[:16].eval()
        for p in vgg.parameters(): p.requires_grad = False
        self.vgg = vgg

    def forward(self, x, y):
        return F.l1_loss(self.vgg(x), self.vgg(y))

#Train
def train(generator, discriminator, dataloader, epochs=10, lr=1e-4):
    g_opt = torch.optim.Adam(generator.parameters(), lr=lr)
    d_opt = torch.optim.Adam(discriminator.parameters(), lr=lr)
    bce = nn.BCEWithLogitsLoss()
    perceptual = VGGPerceptualLoss().to(device)

    generator.train(); discriminator.train()
    generator.to(device); discriminator.to(device)

    for epoch in range(epochs):
        total_g, total_d = 0, 0
        loop = tqdm(dataloader, desc=f"Epoch {epoch+1}")
        for lr, hr in loop:
            lr, hr = lr.to(device), hr.to(device)

            #Dis
            sr = generator(lr).detach()
            d_real = discriminator(hr)
            d_fake = discriminator(sr)
            d_loss = bce(d_real, torch.ones_like(d_real)) + bce(d_fake, torch.zeros_like(d_fake))
            d_opt.zero_grad(); d_loss.backward(); d_opt.step()

            #Gen
            sr = generator(lr)
            d_fake = discriminator(sr)
            content_loss = F.mse_loss(sr, hr)
            adv_loss = bce(d_fake, torch.ones_like(d_fake))
            percept_loss = perceptual(sr, hr)
            g_loss = content_loss + 0.005 * adv_loss + 0.01 * percept_loss
            g_opt.zero_grad(); g_loss.backward(); g_opt.step()

            total_g += g_loss.item(); total_d += d_loss.item()
            loop.set_postfix(G_Loss=g_loss.item(), D_Loss=d_loss.item())

        print(f"Epoch {epoch+1} | G: {total_g/len(dataloader):.4f}, D: {total_d/len(dataloader):.4f}")

#Eval
def evaluate(model, lr_dir, hr_dir, scale=4):
    model.eval()
    psnr_scores, ssim_scores = [], []
    with torch.no_grad():
        for lr_f, hr_f in zip(sorted(os.listdir(lr_dir)), sorted(os.listdir(hr_dir))):
            lr = cv2.imread(os.path.join(lr_dir, lr_f)).astype(np.float32) / 255.
            hr = cv2.imread(os.path.join(hr_dir, hr_f)).astype(np.float32) / 255.

            inp = torch.tensor(lr.transpose(2, 0, 1)).unsqueeze(0).to(device)
            sr = model(inp).cpu().squeeze().numpy().transpose(1, 2, 0)

            # Resize HR to match SR shape (for fair PSNR/SSIM if needed)
            h, w = sr.shape[:2]
            hr_resized = cv2.resize(hr, (w, h), interpolation=cv2.INTER_CUBIC)

            psnr_scores.append(psnr(hr_resized, sr, data_range=1.0))
            ssim_scores.append(ssim(hr_resized, sr, data_range=1.0, channel_axis=2))

    print(f"Avg PSNR: {np.mean(psnr_scores):.2f}, SSIM: {np.mean(ssim_scores):.4f}")


#Visualize
def visualize(model, lr_path, hr_path):
    model.eval()
    lr = cv2.imread(lr_path).astype(np.float32) / 255.
    hr = cv2.imread(hr_path).astype(np.float32) / 255.
    h, w = hr.shape[:2]
    lr_up = cv2.resize(lr, (w, h), interpolation=cv2.INTER_CUBIC)
    inp = torch.tensor(lr_up.transpose(2,0,1)).unsqueeze(0).to(device)
    with torch.no_grad():
        out = model(inp).cpu().squeeze().numpy().transpose(1,2,0)
    plt.figure(figsize=(15,5))
    for i, img in enumerate([lr, out, hr]):
        plt.subplot(1,3,i+1); plt.imshow(img[...,::-1]); plt.axis('off')
    plt.suptitle("LR | SR | HR", fontsize=16); plt.tight_layout(); plt.show()

train_set = SRDataset(train_lr, train_hr, patch_size=48, scale=4)
loader = DataLoader(train_set, batch_size=8, shuffle=True)

gen = Generator(scale=4)
disc = Discriminator()

train(gen, disc, loader, epochs=100)
evaluate(gen, test_lr, test_hr)

# Visualize one sample
visualize(gen, os.path.join(test_lr, os.listdir(test_lr)[0]), os.path.join(test_hr, os.listdir(test_hr)[0]))
```
